## Convolution Operation includes

1. Filter: 

filters are small matrices (e.g., 3x3) that slide over input data (like images), detecting patterns or features. They're learned during training, enabling the network to automatically extract relevant features.

In rgb image we will have a filter of 3 different layer in multidimensional array.

2. Padding Strides : Padding and strides: When performing convolutional operations, the original size of the image changes, potentially leading to loss of information. To address this, we use padding. Strides determine how the filter moves across the image during convolution. For example, a 3x3 filter applies an operation to each row and column, while the stride determines the gap or how many layers to skip between operations. Stride essentially controls the shifting.


3. Pooling Layers

The problem with convolution:
1. Memory issue.: If we take all the values of the image and make a flattening layer, then the matrix will become much more large and complex, so for that, we will use the pooling layer over there.
2. Translation variance: Supposing we are training cats During convolution, features such as ears, nose, and eyes might be identified as belonging to a cat, but their exact locations can vary. If we have a cat in only one place of the image, then it will be called translation variance. To ensure translation invariance, we use pooling layers to downsample the features, making the CNN location-independent for images.

#### Different types of pooling:
- Max pooling: We select a 2x2 matrix (or another specified size) of pixels and retain only the maximum value.
- Min pooling: Similar to max pooling, but we keep the minimum value instead.
- Average pooling: We compute the average value of the pixels in each selected region.
- L2 pooling: We compute the square root of the sum of the squares of the values in each selected region.

#### Advantages of pooling
- It will reduce the size.
- It will help to achieve translation Invariance
- Enhanced Features
- No need of training(pooling layers are the aggregration of the function)

#### Disadvantage of pooling layer
- In image segmentation we can't use this
- Loose of teh information
