# Problem with perceptron

- Perceptron will not work on non-linear data.
- It will not work on XOR operations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions


```

## XOR Gate

```python

data = {
    'Input A': [0, 0, 1, 1],
    'Input B': [0, 1, 0, 1],
    'Output (A XOR B)': [0, 1, 1, 0]
}

df = pd.DataFrame(data)
```

```python
X_train=df.iloc[:,0:2].values
y_train=df.iloc[:,2:3].values.ravel()
```

```python
xor_perceptron=Perceptron()
xor_perceptron.fit(X_train,y_train)
```

```python
plot_decision_regions(X_train,y_train,clf=xor_perceptron)
```

## Non linear data
- let's check perceptron with weak linear relationship

```python
np.random.seed(0)
X = np.random.randn(200, 2)
y = ((X[:, 0] ** 2 + X[:, 1] ** 2) < 1.5).astype(int)
```

```python
dataset=pd.DataFrame(X,y)
```

```python
sns.heatmap(dataset.corr(),annot=True)
```

```python
dataset.corr()
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1],hue=y)
```

```python
non_linear=Perceptron()
non_linear.fit(X,y)
```

```python
plot_decision_regions(X,y,clf=non_linear)
```

- Perceptron will not work on this type of problem where the data is non-linear, even if we run it in an infinite time of epoch or iteration.
