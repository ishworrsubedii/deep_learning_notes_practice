## Loss Function Vs Cost Function

**Loss function is calculated for each sample of single training example and cost function or error function is calculated for entire training set or batch of training set.**

The Mean Squared Error (MSE) loss function is calculated as:

MSE Loss = (yi - ŷi)²

Where:
- yi is the actual value
- ŷi is the predicted value


MSE Cost = 1/N Σ(yi - ŷi)²

Where:
- N is the total number of samples in the entire training set or batch
- yi is the actual value
- ŷi is the predicted value

Here we will covering the following loss functions:
1. Mean Squared Error
- mean square error is the average of the square of the difference between the original values and the predicted values.
- MSE=1/n∑(y−y^)2
- We will do square because we need to add the all the loss or error and ignore the negative values.
- if the error is high for some datapoint then this w,b will be updated in such a way that the error will be reduced.
- Advantages:
    - It is differentiable and convex.
    - 1 minimum point.
- Disadvantages:
    - Error unit is squared.
    - Not robust to outliers.
    - Last layer should be linear.


2. Mean Absolute Error
- mean absolute error is the average of the absolute differences between the original values and the predicted values.
- MAE=1/n∑|y−y^|
- Advantages:
    - Easy to understand.
    - error - same-y (error are represented on real value).
    - Robust to outliers.
- Disadvantages:
    - Not differentiable.(We will use the concept of subgradient)



3. Huber Loss
- If the data has no outlier the it will behave like mse and if the data has outlier then it will behave like mae.
- Hyperparameter delta is used to control the behavior of the loss function.
- Loss function is defined as:
\begin{cases}
0.5 \cdot (y - \hat{y})^2 & \text{for} \ |y - \hat{y}| \leq \delta \\
\delta \cdot |y - \hat{y}| - 0.5 \cdot \delta^2 & \text{for} \ |y - \hat{y}| > \delta
\end{cases} \]

- Here delta is the hyperparameter. If the difference between the actual and predicted value is less than delta then it will behave like mse and if the difference is greater than delta then it will behave like mae.

- It lies in the center between mse and mae.

4. Binary Cross Entropy

- Loss =(-ylog(y^) + (1-y)log(1-y^))
- Cost = 1/n∑(-ylog(y^) + (1-y)log(1-y^))
- Activation function should be sigmoid.
- Advantages:
    - It is differentiable.
    - It is convex.
    - It is robust to outliers.
- Disadvantages:
   - Multiple local minima
   - Hard for intution.

5. Categorical Cross Entropy


\begin{equation}
\text{Categorical Cross-Entropy Loss} = - \sum^{k}_{i} y_i \log(\hat{y}_i)
\end{equation}



- It is used in multiple or multi-class classification.
- where k is the number of classes.
- L=-ylog(y^1)-ylog(y^2)-ylog(y^3)....-ylog(y^k)

6. Sparse Categorical Cross Entropy
- It doesnt have that much different between categorical cross entropy and sparse categorical cross entropy.
- In sparse categorical cross entropy we dont need to convert the target variable into one hot encoding here we can use label encoding for the target variable.
- loss= 
\begin{equation}
\text{Sparse Categorical Cross-Entropy Loss} = - \sum^{k}_{j} y_j \log(\hat{y}_j)
\end{equation}
