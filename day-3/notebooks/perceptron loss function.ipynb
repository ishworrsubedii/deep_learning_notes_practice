```python
import numpy as np

def predict(X, w1, w2, b):
    return np.sign(w1 * X[0] + w2 * X[1] + b)

def train_perceptron(X, y, learning_rate, epochs):
    n = X.shape[0]
    w1 = 1
    w2 = 1
    b = 0

    for epoch in range(epochs):
        misclassified = 0

        for i in range(n):
            prediction = predict(X[i], w1, w2, b)
            if y[i] * prediction <= 0:
                w1 += learning_rate * y[i] * X[i][0]
                w2 += learning_rate * y[i] * X[i][1]
                b += learning_rate * y[i]
                misclassified += 1

        print(f"Epoch {epoch + 1}/{epochs}, Misclassified: {misclassified}/{n}")

        if misclassified == 0:
            break

    return w1, w2, b

# Example usage

learning_rate = 0.1
epochs = 100
w1, w2, b = train_perceptron(X, y, learning_rate, epochs)
print("Final weights:", w1, w2, "Bias:", b)
```
