# Problem with Perceptron Trick

The first problem with the perceptron trick is that it cannot accurately quantify the results.

# Loss Function

The loss function is essentially a mathematical function of weights (`w1` and `w2`) and bias (`b`) in the perceptron. Some examples of loss functions are:

1. **Linear Regression**
   - Mean Squared Error (MSE)
   - Mean Absolute Error (MAE)

2. **Logistic Regression**
   - Binary Cross-Entropy

For the perceptron we typically use the hinge loss for binary classification tasks. If we want to predict probabilities in binary classification, logistic regression with the logistic function and binary cross-entropy loss. For multiclass classification, we can use the softmax function to predict class probabilities and measure the error using categorical cross entropy loss. To minimize the loss we can use gradient descent.

## Hinge Loss for Perceptron

Hinge loss for the perceptron can be defined as:




$f(x_i) = w_1x_{i_1} + w_2x_{i_2} + b$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
```

```python
X,y=make_classification(n_samples=10000,n_redundant=0,n_features=2,n_informative=1,n_classes=2,n_clusters_per_class=1,random_state=68)
```

```python
plt.figure(figsize=(18, 16)) 
custom_palette=['Blue','Red']
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,palette=custom_palette)
plt.show()
```

```python
def perceptron(X,y):
    
    w1=w2=b=1
    lr = 0.01
    
    for j in range(1000):
    
        for i in range(X.shape[0]):

            z = w1*X[i][0] + w2*X[i][1] + b

            if z*y[i] < 0:
                w1 = w1 - lr*y[i]*X[i][0]
                w2 = w2 - lr*y[i]*X[i][1]
                b = b - lr*y[i]
    return w1,w2,b
            
w1,w2,b = perceptron(X,y)      
```

```python
m = -(w1/w2)
c = -(b/w2)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
